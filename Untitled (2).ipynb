```python
import cv2
import os

# 1ï¸âƒ£ Load Haar cascade for face detection
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")

# 2ï¸âƒ£ Folder to save images
folder_path = "Authorized_faces"
if not os.path.exists(folder_path):
    os.makedirs(folder_path)

# 3ï¸âƒ£ Function to capture face images
def capture_faces(person_name, num_samples=50):
    cap = cv2.VideoCapture(0)   # 0 for default webcam
    img_id = 0
    print(f"Starting face capture for {person_name}...")

    while True:
        ret, frame = cap.read()
        if not ret:
            print("Failed to grab frame from webcam")
            break

        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        faces = face_cascade.detectMultiScale(
            gray,
            scaleFactor=1.1,
            minNeighbors=5,
            minSize=(30, 30)
        )

        if len(faces) == 0:
            cv2.putText(frame, "No face detected", (50,50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,0,255), 2)
        else:
            for (x, y, w, h) in faces:
                img_id += 1
                face = gray[y:y+h, x:x+w]
                face = cv2.resize(face, (50,50))

                # Save image
                file_name = os.path.join(folder_path, f"{person_name}.{img_id}.jpg")
                cv2.imwrite(file_name, face)

                # Draw rectangle & counter
                cv2.rectangle(frame, (x,y), (x+w, y+h), (0,255,0), 2)
                cv2.putText(frame, f"Image: {img_id}", (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,0), 2)

        cv2.imshow(f"Capturing {person_name}", frame)

        # Stop if enough images captured or user presses Enter
        if cv2.waitKey(1) == 13 or img_id >= num_samples:
            break

    cap.release()
    cv2.destroyAllWindows()
    print(f"Finished capturing {img_id} images for {person_name}.")

# 4ï¸âƒ£ Example: capture images for Shalu
capture_faces("Shalu", num_samples=50)
```

```python
import cv2
import numpy as np
import os
from random import shuffle

# Folder with authorized face images
folder_path = "Authorized_faces"

# 1ï¸âƒ£ Label function: only one class "Shalu"
def my_label(image_name):
    return np.array([0])  # class 0

# 2ï¸âƒ£ Load dataset
def load_data():
    data = []
    for img in os.listdir(folder_path):
        path = os.path.join(folder_path, img)
        img_data = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
        if img_data is None:
            continue
        img_data = cv2.resize(img_data, (50,50))
        data.append([np.array(img_data), my_label(img)])
    shuffle(data)
    return data

data = load_data()
print("Total images loaded:", len(data))
```

```python
# Convert to arrays
total_images = len(data)
split_index = int(total_images * 0.8)

train = data[:split_index]
test = data[split_index:]

X_train = np.array([i[0] for i in train]).reshape(-1,50,50,1) / 255.0
y_train = np.array([i[1] for i in train])

X_test = np.array([i[0] for i in test]).reshape(-1,50,50,1) / 255.0
y_test = np.array([i[1] for i in test])

print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models

model = models.Sequential([
    layers.Input(shape=(50,50,1)),
    layers.Conv2D(32, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),
    
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),

    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(1, activation='sigmoid')  # binary classification: Shalu or unknown
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.summary()
```

```python
history = model.fit(X_train, y_train, epochs=10, validation_data=(X_test, y_test))
```

```python
model.save("face_recognition_model.h5")
print("Model saved successfully!")
```

```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model

# 1ï¸âƒ£ Load Haar cascade and trained model
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
model = load_model("face_recognition_model.h5")

# 2ï¸âƒ£ Start webcam
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)

    for (x, y, w, h) in faces:
        face = gray[y:y+h, x:x+w]
        face = cv2.resize(face, (50,50))
        face_input = face.reshape(1,50,50,1) / 255.0

        # 3ï¸âƒ£ Predict using the trained model
        prediction = model.predict(face_input)[0][0]

        if prediction > 0.5:
            label = "Shalu - You can enter"
            color = (0,255,0)  # green
        else:
            label = "Unknown - Access Denied"
            color = (0,0,255)  # red

        # Draw rectangle and label
        cv2.rectangle(frame, (x,y), (x+w, y+h), color, 2)
        cv2.putText(frame, label, (x, y-10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.8, color, 2)

    cv2.imshow("Smart Door Access", frame)

    # Press Enter to exit
    if cv2.waitKey(1) == 13:
        break

cap.release()
cv2.destroyAllWindows()
```
